Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:

(более сложный вариант) Если вы будете использовать кроссвалидацию, то преобразования данных и LogisticRegression нужно соединить в один Pipeline с помощью make_pipeline, как это делалось во втором семинаре. Для one-hot-encoding'a можно испльзовать комбинацию LabelEncoder + OneHotEncoder (сначала превращаем строчки в числа, а потом числа првращаем в one-hot вектора.)

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer
```

Не совсем понял, зачем нам LabelEncoder, т.к его использование не лучшая практика, поэтому сделаю pipeline без него

```python
X = data_train.drop(columns='Churn')
y = data_train.Churn

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=44)
```

```python
params = {
    'logisticregression__penalty': ['l1', 'l2'],
    'logisticregression__C': [100, 10, 1, 0.1, 0.01, 0.001],
    'logisticregression__solver': ['liblinear', 'saga']
}

numeric_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown='ignore')

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, num_cols),
        ('cat', categorical_transformer, cat_cols)
    ]
)
model = LogisticRegression()

pipeline = make_pipeline(preprocessor, model)
grid_search = GridSearchCV(estimator=pipeline, param_grid=params, cv=5, 
                           scoring='roc_auc', n_jobs=6, refit=True, 
                           verbose=2, return_train_score=True)

grid_search.fit(X_train, y_train)

print(f"Лучшие параметры: {grid_search.best_params_}")

print(f"Средний результат на кросс-валидации: {grid_search.best_score_}")

predictions = grid_search.predict_proba(X_test)[:, 1]
auc = roc_auc_score(y_test, predictions)
print(f"Результат на тестовой выборке: {auc}")
```

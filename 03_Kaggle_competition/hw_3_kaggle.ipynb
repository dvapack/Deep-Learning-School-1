```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer
```

  ,   LabelEncoder, .     ,   pipeline  

```python
X = data_train.drop(columns='Churn')
y = data_train.Churn

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=44)
```

```python
params = {
    'logisticregression__penalty': ['l1', 'l2'],
    'logisticregression__C': [100, 10, 1, 0.1, 0.01, 0.001],
    'logisticregression__solver': ['liblinear', 'saga']
}

numeric_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown='ignore')

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, num_cols),
        ('cat', categorical_transformer, cat_cols)
    ]
)
model = LogisticRegression()

pipeline = make_pipeline(preprocessor, model)
grid_search = GridSearchCV(estimator=pipeline, param_grid=params, cv=5, 
                           scoring='roc_auc', n_jobs=6, refit=True, 
                           verbose=2, return_train_score=True)

grid_search.fit(X_train, y_train)

print(f" : {grid_search.best_params_}")

print(f"   -: {grid_search.best_score_}")

predictions = grid_search.predict_proba(X_test)[:, 1]
auc = roc_auc_score(y_test, predictions)
print(f"   : {auc}")
```
